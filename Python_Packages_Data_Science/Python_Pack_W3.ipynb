# Introduction to Matplotlib

- Package for data visualization


````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
# Lets generate a dataset
x=np.arange(10)
y=np.arange(0,1000,100)
plt.plot(x,y)
````

````python
# another dataset: an array x containing 100 random numbers drawn from a normal (Gaussian) distribution with a mean of 0.5 and a standard deviation of 0.1.
x=np.random.normal(0.5,0.1,(100))
x.sort() # data sorted from the smalles to the greatest
y=x**2
print(x)
print(y)
plt.plot(x,y)
````

 For the moment we have generated some plots with default settings. For example,we can see that the figures don't have legends or descriptions. Now, let's see how to improve the plots;

````python
# For example we can change the visualizatin  as a scatter plot:
plt.plot(x,y,'o')
````

````python
# Let's create another data set, i,e., 100 numbers from 0 to 99 increasing by 1
x=np.arange(100)
y=x**3
plt.plot(x,y,'.') # an example of another representation
````

````python
# We can also modify the visualization of the line plot
#Alternatives '-' or ':' or '--' or '.-'
plt.plot(x,y,'-.')
plt.show()
plt.plot(x,y,'--')
````

````python
# For the colors: ['r','g','b','c','m','y','k','w']
plt.plot(x,y,'--',color='r')
plt.show()
plt.plot(x,y,'--',color='b')
plt.show()
plt.plot(x,y,'--',color='c')
````

````python
# Here we can also modify the marker size by ms=
plt.plot(x,y,marker='o',ms=10,color='r')
plt.show()
plt.plot(x,y,marker='o',ms=2,color='r')
````

````python
# Here we will combine plots
x=np.arange(0,10,0.5)
y1=x*2
y2=x**2
plt.plot(y1)
plt.plot(y2)
# Notice that here y is being plotted in terms of the index, so it shows 20 values
````

````python
plt.plot(x,y1)
plt.plot(x,y2)
# On the other hand here we plot y in terms of the set values for the variable x
````

````python
# Let's add more elements to the plot
# LABELS
plt.plot(x,y1, label='double')
plt.plot(x,y2, label='square')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
plt.legend() # here we call the previously declared legends
````

````python
plt.plot(x,y1, label='double')
plt.plot(x,y2, label='square')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
plt.legend(['1st','2nd']) # here we can modify the contents of the legend
````

````python
# To compare plots
# subplot creates a grid of plots (rows, columns, position of the subplot [so it changes from 1,2,3, ...n])
plt.subplot(2,1,1)
plt.plot(x,y1, label='double', color='r')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
plt.legend() # now we see the convenience of defining the label beforehand
plt.subplot(2,1,2) # note that the last index changes from 1 to 2 indicating the second plot
plt.plot(x,y2, label='square', color='b')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
plt.legend()
````

````python
# What if we want the two plots, one next to the other ?
plt.subplot(1,2,1)
plt.plot(x,y1, label='double', color='r')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
plt.legend() # now we see the convenience of defining the label beforehand
plt.subplot(1,2,2) # note that the last index changes from 1 to 2 indicating the second plot
plt.plot(x,y2, label='square', color='b')
plt.xlabel('Xs')
plt.ylabel('Ys')
plt.grid()
plt.legend()
````

````python
# Let's create a plot using manually provided data points
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.plot(x,y,'o')
````

````python
# Let's use another option to plot the individual data points using the scatter plot
plt.scatter(x,y)
````

````python
# We can create also a colormap which is like adding a third dimension fo the plot
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors=np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
plt.scatter(x,y,c=colors,cmap='jet')
plt.colorbar()
````

````python
# We can do something similar with the sizes
x=np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y=np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes=np.array([100,500,200,30,400,45,500,550,60,70,80,90,100])
plt.scatter(x,y,s=sizes)
````

````python
# We can combine the sizes and the colors also
# alpha controls the color intensity
plt.scatter(x,y,c=colors, cmap='jet',alpha=0.5,s=sizes)
plt.colorbar()
````

````python
# let's see another plot type
plt.bar(x,y)
````

````python
# let's see a pie chart
x=[0.66,0.25,0.09]
plt.pie(x,labels=['2/3','1/5','others'])
plt.legend()
````

````python
# explode argument helps to empasize the data
# notice here that explode has three values associated to each data set. The difference is given by the second value which is affected by the explode command
x=[0.66,0.25,0.09]
plt.pie(x,labels=['2/3','1/5','others'], explode=[0.0,0.1,0])
plt.legend()
````

````python
# explode argument helps to empasize the data
# notice here that explode has three values associated to each data set. The difference is given by the second value which is affected by the explode command
x=[0.66,0.25,0.09]
plt.pie(x,labels=['2/3','1/5','others'], explode=[0.0,0.2,0.1])
plt.legend()
````

````python
# Here we can compare different plots
x=np.arange(10)
alpha=(x+1)**2/2
delta=(x+2)**2
omicron=(x+3)**2*3

plt.plot(x,alpha,label='Alpha')
plt.plot(x,delta,label='Delta')
plt.plot(x,omicron,label='Omicron')
plt.legend()
````

## Play with other plots
You can play with other plots we learned here. Be as creative as you can and see how the visualization helps you to understand the data better.

````python
# Here we will use the colormap to add another "dimension", the species
# To personalize the labels we couls use mpatches
# I ised this tutorial to add the legends based on colors https://www.tutorialspoint.com/how-to-manually-add-a-legend-with-a-color-box-on-a-matplotlib-figure

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
species_colors = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
lines=plt.scatter(data['sepal_length'], data['petal_length'], c=data['species'].map(species_colors), alpha=0.5)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.grid()

# Create a legend with a color box
setosa_patch = mpatches.Patch(color='red', label='Setosa')
versicolor_patch = mpatches.Patch(color='green', label='Versicolor')
virginica_patch = mpatches.Patch(color='blue', label='Virginica')
plt.legend(handles=[setosa_patch, versicolor_patch, virginica_patch], loc='upper left', framealpha=0.2, frameon=True)
````
