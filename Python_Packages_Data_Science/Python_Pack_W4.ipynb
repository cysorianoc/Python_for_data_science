## Introduction to Seaborn

- To plot different types of graphs.

````python
# First at all: LOAD DATA AND IMPORT PACKAGES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Also import one of seaborn datasets
tips=sns.load_dataset('tips')
tips.head()
````

````python
# We can print the loaded dataframe to see more details
# total bill is the amount paid by a customer
# And there are some other aatributes from data at a restaurant
tips
````

## Types of plots:
- Relational Plots
- Distribitional Plots
- Categorical Plots
- Joint Plots
- Pair Plots

````python
# Relational Plots
sns.relplot(data=tips,x='total_bill',y='tip')
````

````python
# Here hue='sex' helps to make a differentiation of the data, is like adding an extra dimension to the plot
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex')
````

````python
# same for the day
sns.relplot(data=tips,x='total_bill',y='tip',hue='day')
````

````python
sns.relplot(data=tips,x='total_bill',y='tip',hue='time')
````

````python
# We can sepparate plots and keep them side by side by using col as sepparator:
sns.relplot(data=tips,x='total_bill',y='tip',hue='smoker',col='time')
````

````python
# In this example as we have several sizes, we can create something similar to a colormap
sns.relplot(data=tips,x='total_bill',y='tip',hue='size')
````

````python
# Here we can add a size to the data points:
sns.relplot(data=tips,x='total_bill',y='tip',size='size',hue='size')
````

````python
# We can also provide a range for the sizes to create a contrast:
sns.relplot(data=tips,x='total_bill',y='tip',size='size',sizes=(15,200),hue='size')
````

````python
# The parameter alpha here provides transparency to the data points:
sns.relplot(data=tips,x='total_bill',y='tip',size='size',sizes=(15,200),alpha=0.5,hue='size')
````

````python
# Let's see also how the days and the total bill are related:
sns.relplot(data=tips,x='day', y='total_bill')
````

````python
# We can see that also with lines
sns.relplot(data=tips,x='day', y='total_bill', kind='line')
````

````python
sns.relplot(data=tips,x='day', y='total_bill', hue='sex',kind='line')
````

````python
sns.relplot(data=tips,x='day', y='total_bill', hue='sex',kind='line', col='time')
````

## JointPlots
We can do multiple types of plots together

````python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
````

````python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='scatter')
````

````python
# We can visualize the different types of plots:
kind=['scatter', 'hist','hex', 'kde', 'reg', 'resid']
for k in kind:
  sns.jointplot(data=tips, x='total_bill', y='tip', kind=k)
````

````python
# We can plot the variables with good correlation observed from the colormap:
sns.jointplot(data=data, x='sepal_length', y='petal_length', kind='reg')
````

````python
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame and has been previously loaded
x = data['sepal_length']
y = data['petal_length']

# Compute the Pearson correlation coefficient and the p-value
r, p = stats.pearsonr(x, y)
r2_value = r ** 2  # Compute R^2 value

# Create a jointplot with regression line
jointplot = sns.jointplot(x=x, y=y, kind="reg")

# Annotate the plot with the R^2 value using plt.annotate
# Adjust the position and formatting to your preference
plt.annotate(f'$R^2 = {r2_value:.2f}$', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12)
plt.annotate(f'$R = {r:.2f}$', xy=(0.05, 0.90), xycoords='axes fraction', fontsize=12)
plt.show()
````

````python
sns.relplot(data=data,x='sepal_length',y='petal_width', col='species')
````

````python
# Here we will see how to plot filetred data
# Assuming 'data' is your DataFrame and it's already loaded
# Filter the DataFrame for rows where the species is 'setosa'
setosa_data = data[data['species'] == 'setosa']

# Now, plot only the setosa data using relplot without using 'col' for species differentiation
sns.relplot(data=setosa_data, x='sepal_length', y='petal_length')
````
