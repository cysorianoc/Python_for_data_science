## Calculate Payment
*This is an example of the hierarchical view of a program*
- Create a program to calculate net pay for a paycheck:
  - The pay will be calculated based on hours and rate;
  - gross pay=hours*rate
  - overtime pay=(hours-40)x rate x 0.5+(hours-80)x rate.
  - tax=(total_pay-500)*20%
  - benefit=total_pay*15%
  - withholding=tax+benefit
  - net_pay=total pay- withholdings






```python
# to build the program, we start from the bottom of the chart,
# from the get_inputs

def get_hours_worked():
  return int(input('Enter the number of hours worked:'))

def get_hourly_rate():
  return int(input('Enter the hourly rate of work:'))

def get_inputs():
  return get_hours_worked() , get_hourly_rate()

# now we move to the calculations
def calc_gross_pay(hours, rate):
  return hours*rate

def calc_overtime(hours,rate):
  if hours<40:
    return 0
  elif hours <80:
   return (hours-40)*rate*0.5
  else:
    return (hours-40)*rate*0.5+(hours-80)*rate

# now we move to the withholding, taxes and benefits

def calc_tax(total):
  if total<500:
    return 0
  else:
    return (total-500)*0.2

def calc_benefit(total):
  return total*0.15


def calc_withholding(total):
  return calc_tax(total)+calc_benefit(total)

  # now we move to the total pay

def calc_net_pay(total, withholding):
  return total-withholding

def main():
  hours,rate=get_inputs()
  gross=calc_gross_pay(hours,rate)
  overtime=calc_overtime(hours,rate)
  total=gross+overtime
  withholdings=calc_withholding(total)
  return calc_net_pay(total,withholdings)
```

```python
main()
```

## Introduction to Recursion
- Recursive function: a fucntion that calls itself.
- Recursion is a powerful tool for solving repetitive problems.
- Recursion is never required to solve a problem.
  - Any problem that can be solved recursively can be solved with loops.
  - Recursive algorithms usually less efficient than iterative ones: due to overhead of each function call.

- Recursion function must have a way to control the number of times it repeats.
  - usually involves an ``if-else`` statement which defines when the fucntion should return a value when it should call itself.
- Depth of recursion: the number of times a function calls itself.

### General Outline
- If the problem can be solved now without recursion, solve and return.
  - known as *base case*
  - otherwise, reduce the problem into a smaller problem of the same structure and call the function again to solve the smaller problem
  - Known as the recursive case

### n! problem
- In mathematics, the $n!$ notation represents the factorialk of a number n
  - 0!=1
  - $n!=n*(n-1)*(n-2)*...*2*1$
- The above definition lends itself to recursive programming.
  - n=0 is the base case, return 1
  - n>0 is the recursive case, call itself:
    - ``factorial (n)=n*factorial(n-1)``




```python
# now let's put that on practice
# f(n)=n*(n-1)*(n-2)*...*3*2*1
# f(0)=1
# f(n)=n*f(n-1)

def f(n):
  if n==0:
    return 1
  else:
    return n*f(n-1)
```

```python
f(4)
```

```python
f(5)
```

## Using Recursion
- Since each call to the recursive function reduces the problem:
  - Eventually, it will get to the base case which does not require recursion, and the recursion will stop.
- Usually the problem is reduced by making one or more parameters smaller at each function call

## Direct and indirect recursion
- Direct recursion: when a function directly calls itself.
- Indirect recursion: when function A calls Function B, which in turn calls function A.

## Recursion vs Looping
- Reasons not to use recursion:
  - Less efficient: entails function calling overhead thst is not necessary with a loop.
  - Usually a solution using a loop is more evident than using a recursive function.
- Some problems are more easily solved with recursion than with a loop.
  - Example: Fibonacci, where the mathematical definition lends itself to recursion.

###Task: define a recursive function to compute fibonacci series: $fib(0) = 0$, $fib(1) = 1$, $fib(n) = fib(n-1) + fib(n-2)$

```python
def fib(n):
  if n==0:
    return 0
  elif n==1:
    return 1
  else:
    return fib(n-1)+fib(n-2)
```

```python
for i in range(10):
  print(fib(i))
```

###Task: define a recursive function to compute the number of digits of a positive integer.

```python
"""The function digits(123) is called.
Since 123 is not less than 10, it goes to the else block.

It calls digits(12) with n/10, which is effectively 12.

Now, digits(12) is called.

Since 12 is still not less than 10, it goes to the else block again.

It calls digits(1) with n/10, which is effectively 1.

Now, digits(1) is called.

1 is less than 10, so it returns 1.

Now, we go back up the recursion chain:
digits(1) returned 1, so in digits(12), we have 1 + 1 = 2.
digits(12) returned 2, so in digits(123), we have 1 + 2 = 3.
Finally, digits(123) returns 3.

So, the result of digits(123) is 3, indicating that there are three digits in 123.
The function counts the number of digits by recursively dividing the number by 10 until it becomes less than 10, adding 1 for each division."""

def digits(n):
  if n<10:
    return 1
  else:
    return 1+digits(n/10)

digits(123)

```

### Task: define a function to check whether or not a string without special characters is palindrome.

Definition: Palindrome is a word, phrase, or sequence that reads the same backward as forward.

For example:
*   '', 'a', 'ada', 'adda' are palindrome;
*   'ab', 'abs' are not palindrome.

We can assume the input string contains only A to Z and a to z, no other special characters, it might be empty, and an empty string is palindrome.

Hint: The first character of the string *s* can be accessed by *s[0]* and the last character of the string *s* can be accessed by *s[-1]*.
The substring without the first and last character of a string *s* can be obtained by *s[1:-1]*

```python
def palindrome(s):
  if len(s)==0 or len(s)==1:
    return True
  else:
    if s[0]==s[-1]:
      return True
    else:
      return False

print(palindrome('radar'))
print(palindrome('adda'))
print(palindrome('ads'))
```
