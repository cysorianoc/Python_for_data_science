### Void or Value-Returning
- A void function: simply executes the statements it contains and it terminates. For instance ``print()`` once the line is printed, it is terminated. It executes or shows a functionallity and thats it.
- A value-returning function: for instance ``input``. So it executes the statements it containes, and then it returns a value back to the statement that called it.

### Parameters
- A function with no parameters:
  - simply execute the statement inside of the function without inputs.
- A function with parameters:
  - Execute the statements with extra inputs passed to these parameters.

### Functions in functions
- Hierarchical functions:
  - Break down a big tast into small subtasks, then break down smaller subtasks. Used in object-orientyed project, where a project is subdivided in tasks.
- Recursive functions:
  - A function calls itself directly or indirectly.

  ### A Description of a Function
- To describe a function, you need three elements:
  - Name: so you can call it.
  - Inputs: the arguments to call it.
  - Outputs: what outputs does this function produce ?

### Function name
- Function name should be descriptive of the task carried out by the function.

### Code Block
- A function is a block of code
- Each block must be indented
  - Lines in block must begin with the same number of spaces
    - Use tabs or spaces to indent lines in ablock.
    - IDLE (Python's Integrated Development and Learning Environment) automatically indents the lines in a block.
- Black lines that appear in a block are ignored.

### Calling a Function

- Call a function to execute it
  - Just call the name of the function
  - If the function needs arguments, you should feed it with proper arguments.
- When a function is called:
 - The flow jumps to the function and executes statements in the block.
 - The flow jumps back to part of program that called the function, also known as function return.

One advantage in the definition of functions is if we want to change something in our code, let's say, instead of saying 'Hello', say 'Hi' and repeat the message ten times. We can do it directly in the function and when calling it this will be reflected immediately.

```python
# goodbye(): calling this function will print a count down from 10 to 1, then print the message "good bye!"
def goodbye():
  n=11
  for i in range(11):
    if i > 0:
      n=n-1
      print(n)
  print('good bye')
```

```python
goodbye()
```

```python
# another solution
def goodbye2():
  count=10
  while (count>0):
    print(count)
    count-=1
  print('good bye')
```

```python
goodbye2()
```

###Task: define a function that print primary integers in 100 (hint: you can use loops)

```python
# your code will be here
def prime100():
    for i in range (1,101):
        prime=True
        for j in range(2,i):
            if i%j==0:
              prime=False
              break
        if prime:
          print(i)
```

```python
prime100()
```

```python
# solution
def prime_in_100():
  for i in range(1, 100):
    prime = True
    for j in range(2, i):
      if i % j == 0:
        prime = False
        break
    if prime:
      print(i)

prime_in_100()
```

###Task: define a function that print even number from 1 to 10.

```python
def evens():
  counter=0
  for i in range(10):
    counter+=1
    if counter%2==0:
      print(counter)
```

```python
evens()
```

```python
# solution
def even_in_10():
  for i in range(1, 11):
    if i % 2 == 0:
      print(i)

even_in_10()
```

###Task: define a function that print odd number from 1 to 10

```python
def odds():
  counters=0
  for i in range(10):
    counters+=1
    if counters%2!=0:
      print(counters)
```

```python
odds()
```

```python
# solution
def odd_in_10():
  for i in range(1, 11):
    if i % 2 != 0:
      print(i)

odd_in_10()
```
