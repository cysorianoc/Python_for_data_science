## What are Parameters ?
- Define functions with parameters
- Define functions with multiple parameters
- Call functions with one argument

## Parameters
- The values needed to the function to work.
- When defining a function, parameters are placed in parenthesis following the function name.
  - Parameters are used as variables inside the function.

## Defining a function with parameters
- Function definition: specifies what function does
- ``def function_name(parameter list):``\
``statement``\
``statement``\
- Function header: first line of function definition.
  - ``def`` indicates the beginning of function definition.
  - ``function_name`` specifies the name of the function.
  - end with a ``:``

## Arguments
- The values we place in the parenthesis when calling the function.
- Arguments are pieces of data that are sent into the functions.
- Parameters will be the name of variables that will work in the statements inside the function.

# What are multiple parameters ?
- Python allows writing functions that accept multiple arguments.
  - The parameter list replaces single paremeter and each parameter is separated by comma.
- Arguments are passed by position to corresponding parameters
  - First parameter receives values of first arguments, second receives the value of second argument and so on.
- Types of function arguments: keywords, default values, variable length parameters.

So we call the parameters inside the function either by position or by keywords.

Now let's see some examples:

```python
# define a function sum(x,y) that takes two numbers and print x+y
def sum(x,y):
  print(x+y)
```

```python
# for example:
sum(3,2)
```

```python
# define a function times(x,y) that takes two numbers and print x*y
def times(x,y):
  print(x*y)
```

```python
times(5,3)
```

```python
# define a function times(x,y) that takes two numbers and print x*y but this time you cannot use * operation (hint: 3*4=3+3+3+3)
def times(x,y):
  sum=0
  for i in range(y):
    sum=x+sum
  print(sum)
```

```python
times(-1,10)
```

```python
def times(x,y):
  sum=0
  for i in range(y):
    sum+=x
  print(sum)
```

```python
times(2,3)
```

Note that the previous code will fail if we use a negative number in the second argument. So we need to emphazyse that the second parameter cannot be negative. Let's see an example:

```python
times(1,-1)
```

```python
# with emphasys:
times(y=1,x=-1)
```

```python
# define a function remainder(n,m) that print the remainder n%m
def remainder(n,m):
  print(n%m)
```

```python
remainder(15,3)
```

```python
# by keywords
remainder(n=15, m=3)
```

```python
# we can switch the position of the inputs
remainder(m=3,n=15)
```

## Calling functions with keyword arguments

- Keyword argument: argument that specifies which parameter the value should be passed to;
  - Position when calling the function is irrelevant.
  - General format:
  - ``function_name(parameter=value)``
- Possible to mix keyword and positional arguments when calling a function.
  - Positional arguments must appear first.


## Making changes to Parameters

- Changes made to a parameter value within the function do not affect the argument.
  - Known as pass by value.
  - Provides a way for unidirectional communication between one function and another function.
  - Calling a function can communicate with called function.
